```python
def translate(text, src, tgt):
    try:
        result = translator(
            text,
            src_lang=LANGUAGE_MAP[src],
            tgt_lang=LANGUAGE_MAP[tgt]
        )
        return result[0]['translation_text']
    except Exception as e:
        return f"❌ Error: {str(e)}"
```

```python
with gr.Blocks() as app:
    gr.Markdown("# 🌍 NLLB-200 Translator (204 Languages)")

    src = gr.Dropdown(list(LANGUAGE_MAP.keys()), label="From Language")
    tgt = gr.Dropdown(list(LANGUAGE_MAP.keys()), label="To Language")

    text = gr.Textbox(label="Enter text")
    out = gr.Textbox(label="Translated Text")

    btn = gr.Button("Translate")
    btn.click(translate, inputs=[text, src, tgt], outputs=out)

app.launch()
```
